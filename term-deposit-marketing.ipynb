```python
from sklearn.preprocessing import LabelEncoder

LE = LabelEncoder()
data['job'] = LE.fit_transform(data['job'])
data
```

```python
data.month = data.month.replace({"jan": "January",
                                                   "feb": "February",
                                                   "mar":"March",
                                                   "apr":"April",
                                                   "may":"May",
                                                   "jun":"June",
                                                   "jul":"July",
                                                   "aug":"August",
                                                   "oct":"October",
                                                   "nov":"November",
                                                   "dec":"December"})
```

```python
data['month'] = pd.to_datetime(data.month, format='%B').dt.month
```

```python
data['month'] = pd.to_datetime(data.month, format='%d').dt.month
data
```

```python
data.y.value_counts()
data['y'] = data['y'].map(lambda x: 0 if x=="no" else 1)
data.y.value_counts()
```

```python
from sklearn.model_selection import train_test_split


train_valid_X = data[data.columns.difference(['y'])]
train_valid_y = data.y
train_X , valid_X , train_y , valid_y = train_test_split( train_valid_X , train_valid_y , train_size = .8 )


print (data.shape , train_X.shape , valid_X.shape , train_y.shape , valid_y.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier , GradientBoostingClassifier

model0 = RandomForestClassifier(n_estimators=100)
model1 = SVC()
model2 = GradientBoostingClassifier()
model3 = KNeighborsClassifier(n_neighbors = 3)
model4 = GaussianNB()
model5 = LogisticRegression()

model0.fit( train_X , train_y )
model1.fit( train_X , train_y )
model2.fit( train_X , train_y )
model3.fit( train_X , train_y )
model4.fit( train_X , train_y )
model5.fit( train_X , train_y )

print ("Random Forest:", model0.score( train_X , train_y ) , model0.score( valid_X , valid_y ))
print ("SCV:", model1.score( train_X , train_y ) , model1.score( valid_X , valid_y ))
print ("Gradient Boost:", model2.score( train_X , train_y ) , model2.score( valid_X , valid_y ))
print ("K Neighbors:", model3.score( train_X , train_y ) , model3.score( valid_X , valid_y ))
print ("Gaussian:", model4.score( train_X , train_y ) , model4.score( valid_X , valid_y ))
print ("Logical Reg:", model5.score( train_X , train_y ) , model5.score( valid_X , valid_y ))
```

```python
from numpy import mean
from numpy import std
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

cv = KFold(n_splits=5, random_state=1, shuffle=True)

scores = cross_val_score(model0, train_X, train_y, scoring='accuracy', cv=cv, n_jobs=-1)
print('Random Forest 5-fold Cross Validation Accuracy: %.3f (%.3f)' % (mean(scores), std(scores)))
scores = cross_val_score(model1, train_X, train_y, scoring='accuracy', cv=cv, n_jobs=-1)
print('SCV 5-fold Cross Validation Accuracy: %.3f (%.3f)' % (mean(scores), std(scores)))
scores = cross_val_score(model2, train_X, train_y, scoring='accuracy', cv=cv, n_jobs=-1)
print('Gradient Boost 5-fold Cross Validation Accuracy: %.3f (%.3f)' % (mean(scores), std(scores)))
scores = cross_val_score(model3, train_X, train_y, scoring='accuracy', cv=cv, n_jobs=-1)
print('K Neighbors 5-fold Cross Validation Accuracy: %.3f (%.3f)' % (mean(scores), std(scores)))
scores = cross_val_score(model4, train_X, train_y, scoring='accuracy', cv=cv, n_jobs=-1)
print('Gaussian 5-fold Cross Validation Accuracy: %.3f (%.3f)' % (mean(scores), std(scores)))
scores = cross_val_score(model5, train_X, train_y, scoring='accuracy', cv=cv, n_jobs=-1)
print('Logical Reg 5-fold Cross Validation Accuracy: %.3f (%.3f)' % (mean(scores), std(scores)))
```
